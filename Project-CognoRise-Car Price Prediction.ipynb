### DATA MODELLING

```python
X=df.drop(columns='price')
Y=df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.20, random_state=42)
```

```python
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['CarName','drivewheel','cylindernumber']])
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['CarName','drivewheel','cylindernumber']), remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train, Y_train)
```

```python
Y_pred=pipe.predict(X_test)
```

```python
Y_pred
```

```python
r2_score(Y_test, Y_pred)
```

```python
score=[]
for i in range(1000):
    X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.20, random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train, Y_train)
    Y_pred=pipe.predict(X_test)
    score.append(r2_score(Y_test, Y_pred))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.20, random_state=np.argmax(score))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train, Y_train)
Y_pred=pipe.predict(X_test)
print(r2_score(Y_test, Y_pred))
```
